## Calculate LDA for different numbers of topics

```python
texts = [feature.split(' ') for feature in features]
id2word = gensim.corpora.Dictionary(texts)
```

```python
corpus = [id2word.doc2bow(text) for text in texts]
lst = [[(id2word[id], freq) for id, freq in cp] for cp in corpus[:1]]
```

```python
len(id2word)
```

```python
models = []

for i, no_topics in enumerate([40]):
    lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                               id2word=id2word,
                                               num_topics=no_topics, 
                                               random_state=0,
                                               update_every=0,
                                               alpha='auto',
                                               per_word_topics=True,
                                               passes=passes)
    print('finished model')
    
    # Compute Perplexity
    perplexity = math.exp(lda_model.log_perplexity(corpus))
    print('{0}\nPerplexity: {1}'.format(no_topics, perplexity))  # a measure of how good the model is. lower the better.
    models.append({
        'no_topics': no_topics,
        'perplexity': perplexity,
        'model': lda_model
    })
    
    print(lda_model.print_topics(num_topics=no_topics))

    # Compute Coherence Score
    #coherence_model_lda = gensim.models.CoherenceModel(model=lda_model, texts=data_lemmatized, dictionary=id2word, coherence='c_v')
    #coherence_lda = coherence_model_lda.get_coherence()
    #print('\nCoherence Score: ', coherence_lda)
    
```

# Analyze
## Function for displaying

```python
def get_topics(H, W, feature_names, chapters, no_top_words, no_top_documents):
    topics = {
        'id': [],
        'word': []
    }
    for topic_idx, topic in enumerate(H):
        topics['id'].append(topic_idx)
        topics['word'].append(" ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
    return pd.DataFrame(topics)
```

## Save

save the 5 most important topics

```python
def find_highest_five(nmf_W):
    highest_five = []
    for i, w in enumerate(nmf_W):
        found = False
        for h in highest_five:
            if w >= h['value']:
                highest_five.append({
                    'index': i,
                    'value': w
                })
                found = True
                break
        if len(highest_five) < 5 and not found:
            highest_five.append({
                'index': i,
                'value': w
            })
        highest_five = sorted(highest_five, key=lambda x: x['value'], reverse=True)[0:5]
    return highest_five
```

```python
l = []
for w in nmf_W:
    l.append(find_highest_five(w))
nmf_df = pd.DataFrame(l)
nmf_df.columns = ['topic_1', 'topic_2', 'topic_3', 'topic_4', 'topic_5']
nmf_df.insert(0, 'features', features)
nmf_df = pd.concat([df_chapters.iloc[:,0:4], nmf_df], axis=1)
nmf_df.head()
```
