```python
import sys
sys.path.append("..")

from services.segments_database import do_query, insert_trainingdata_from_chapter
import pandas as pd
```

```python
current_category = 'zk'
zk_category_id = 4
```

```python
results = []
query = '''
  SELECT *
  FROM chapter
  ORDER BY RAND()
  LIMIT {};
'''

res = do_query(query.format(10))
df = pd.DataFrame(res.fetchall())
df.columns = res.keys()
results.append(df)
    
df = pd.concat(results, ignore_index=True)
df.shape
```
